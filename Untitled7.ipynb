```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "meta-llama/Llama-3.1-8B-Instruct"  # use Instruct for Q&A

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    dtype=torch.bfloat16,   # or torch.float16 if bfloat16 not supported
    load_in_4bit=True       # saves VRAM, works on T4/V100 GPUs in free Colab
)
```

```python
question = "what is AI and AI Agent"

inputs = tokenizer(question, return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7,
    top_p=0.9
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
